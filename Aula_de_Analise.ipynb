```python
# passo 0 - Entender o desafio
# passo 1 - Percorrer todos os arquivos da pasta de base de dados


```

```python
import os #para mexer em pastas de arquivos no pc, etc
import pandas as pd
lista_arquivo = os.listdir("/content/drive/MyDrive/Vendas")
display(lista_arquivo)
```

```python
tabela_total = pd.DataFrame()  #criar uma tabela vazia, vamos juntar os documentos todos em apenas 1

# passo 2 - Importar a base de dados de vendas
for arquivo in lista_arquivo:
#se tem "vendas" dentro do arquivo se usar o not antes de vendas mostra tudo o resto sem vendas
  if "Vendas" in arquivo:  #arquivo.lower() ele ia mostrar com as maiusculas e minusculas
   # print(f"/content/drive/MyDrive/Vendas/{arquivo}")  #formatar uma variavel dentro dele usar o f
    tabela = pd.read_csv(f"/content/drive/MyDrive/Vendas/{arquivo}")
    #print(arquivo)
    #display(tabela)
    tabela_total = tabela_total.append(tabela)  #adicionamos a nossa tabela à tabela criada por nos em cima

# passo 3 - Tratar/ compilar base de dados

display(tabela_total)
```

```python
# passo 4 - Calcular o produto mais vendido em (quantidade),  agrupar os produtos e somar a quantidade vendida .sum para juntar os valores das restantes tabelas

tabela_produtos = tabela_total.groupby("Produto").sum()  
tabela_produtos = tabela_produtos[["Quantidade Vendida", "Preco Unitario"]].sort_values(by="Quantidade Vendida", ascending =False)
#sort_values para ordenar por valores, ascending ordem ascendente
display(tabela_produtos)
```

```python
# passo 5 - Calcular o produto que mais faturou
#criou se uma coluna de faturamento, ao multiplicar os valores da quantidade vendida e preco unitario
tabela_total["Faturamento"] = tabela_total["Quantidade Vendida"] * tabela_total["Preco Unitario"]
#display(tabela_total)

tabela_faturamento = tabela_total.groupby("Produto").sum()
tabela_faturamento = tabela_faturamento[["Faturamento"]].sort_values(by="Faturamento", ascending =False)

display(tabela_faturamento)
```

```python
# passo 6 - Calcular a loja/cidade que mais vendeu (em faturamento) - criar gráfico/dashboard
tabela_lojas = tabela_total.groupby("Loja").sum()
tabela_lojas = tabela_lojas[["Faturamento"]]#.sort_values(by="Faturamento", ascending=False)
display(tabela_lojas)
```

```python
import plotly.express as px
```

```python
grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y="Faturamento")  #o nome "Loja" é o indice da tabela(esta a negrito) ou seja temos de chamar o index
grafico.show()
```
